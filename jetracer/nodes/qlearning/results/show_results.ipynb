```python
import h5py
import os

h5_path = 'data/replay_buffer_205746.h5'
if not os.path.exists(h5_path):
    print(f'Plik {h5_path} nie istnieje!')
else:
    with h5py.File(h5_path, 'r') as f:
        print(f"{'Nazwa':<20} {'Kształt':<20} {'Typ danych':<15}")
        print('-'*60)
        for key in f.keys():
            shape = f[key].shape
            dtype = f[key].dtype
            print(f"{key:<20} {str(shape):<20} {str(dtype):<15}")
```
